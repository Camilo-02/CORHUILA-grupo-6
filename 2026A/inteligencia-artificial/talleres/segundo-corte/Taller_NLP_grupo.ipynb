## **Ejercicio 1: Limpieza y Preprocesamiento Básico**

**Objetivo:** Crear una función que limpie texto removiendo URLs, emails, caracteres especiales y normalizando espacios.

**Instrucciones:**
1. Implementa la función `limpiar_texto()` que:
   - Convierte a minúsculas
   - Remueve URLs (http, https, www)
   - Remueve emails
   - Remueve caracteres especiales manteniendo solo letras y espacios
   - Normaliza espacios en blanco

2. Prueba con los siguientes textos:

```python
import re

def limpiar_texto(texto):
    """
    Limpia un texto removiendo elementos no deseados.

    Args:
        texto: String con el texto a limpiar

    Returns:
        String con el texto limpio
    """

    # 1. Convertir a minúsculas
    texto = texto.lower()

    # 2. Remover URLs (http, https, www)
    texto = re.sub(r'https?://\S+|www\.\S+', '', texto)

    # 3. Remover emails
    texto = re.sub(r'\S+@\S+', '', texto)

    # 4. Remover caracteres especiales (solo letras y espacios)
    texto = re.sub(r'[^a-záéíóúüñ\s]', '', texto)

    # 5. Normalizar espacios (eliminar espacios múltiples)
    texto = re.sub(r'\s+', ' ', texto).strip()

    return texto

# Casos de prueba
textos_prueba = [
    "¡Hola!!! Visita https://ejemplo.com o envía tu CV a recurso@empresa.com. Hay   espacios extra!!!",
    "Python3.9 es excelente!!! Descarga desde www.python.org #programming @user",
    "El NLP es fascinante... ¿No lo crees? (sí/no) [responder]"
]

print("Resultados de limpieza:")
print("=" * 60)
for i, texto in enumerate(textos_prueba, 1):
    print(f"\nTexto {i}:")
    print(f"Original:  {texto}")
    print(f"Limpio:    {limpiar_texto(texto)}")
```

## **Ejercicio 2: Tokenización**

**Objetivo:** Dividir textos en tokens (palabras u oraciones) utilizando NLTK.

**Instrucciones:**
1. Descarga los recursos necesarios de NLTK
2. Implementa tokenización de palabras y oraciones
3. Compara los resultados para diferentes textos

```python
import nltk

# Descargar recursos necesarios
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize, sent_tokenize

def analizar_tokenizacion(texto, idioma='spanish'):
    """
    Realiza tokenización de un texto en palabras y oraciones.
    """

    # Tokenización por oraciones
    tokens_oracion = sent_tokenize(texto, language=idioma)

    # Tokenización por palabras
    tokens_palabra = word_tokenize(texto, language=idioma)

    # Crear resultado
    resultado = {
        'tokens_palabra': tokens_palabra,
        'tokens_oracion': tokens_oracion,
        'num_palabras': len(tokens_palabra),
        'num_oraciones': len(tokens_oracion)
    }

    return resultado

# Casos de prueba
texto1 = "Hola. ¿Cómo estás? Estoy muy bien, gracias por preguntar."
texto2 = "El procesamiento del lenguaje natural es increíble! ¿No crees? Sí, totalmente."

for i, texto in enumerate([texto1, texto2], 1):
    print(f"\nTexto {i}: {texto}")

    resultado = analizar_tokenizacion(texto)

    print("\nTokens de oraciones:")
    print(resultado['tokens_oracion'])

    print("\nTokens de palabras:")
    print(resultado['tokens_palabra'])

    print("\nNúmero de palabras:", resultado['num_palabras'])
    print("Número de oraciones:", resultado['num_oraciones'])
```

## **Ejercicio 3: Normalización de Texto**

**Objetivo:** Normalizar texto removiendo acentos y caracteres diacríticos.

**Instrucciones:**
1. Implementa función que remueva acentos usando `unicodedata`
2. Convierte textos con acentos a su forma normalizada
3. Compara el antes y después

```python
import unicodedata

def normalizar_texto(texto):
    """
    Normaliza un texto removiendo acentos y caracteres diacríticos.
    """

    # 1. Convertir a minúsculas
    texto = texto.lower()

    # 2. Normalizar a forma NFD (separa letras de acentos)
    texto = unicodedata.normalize('NFD', texto)

    # 3. Filtrar caracteres diacríticos (categoría 'Mn')
    texto = ''.join(
        c for c in texto
        if unicodedata.category(c) != 'Mn'
    )

    return texto

# Casos de prueba
textos_acentuados = [
    "¡Hola! ¿Cómo Estás?",
    "Programación y árboles de decisión",
    "La física y química de la Tierra",
    "Café con leche es delicioso"
]

print("Normalización de textos acentuados:")
print("=" * 60)
for texto in textos_acentuados:
    print(f"Original:     {texto}")
    print(f"Normalizado:  {normalizar_texto(texto)}")
    print()
```

## **Ejercicio 4: Remoción de Stopwords**

**Objetivo:** Remover palabras comunes que no aportan significado.

**Instrucciones:**
1. Descarga la lista de stopwords en español
2. Implementa función que filtre stopwords
3. Analiza cómo cambia un texto después de remover stopwords

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def remover_stopwords(texto, idioma='spanish'):
    """
    Remueve stopwords de un texto.
    """

    # 1. Obtener lista de stopwords
    stop_words = set(stopwords.words(idioma))

    # 2. Tokenizar el texto
    palabras = word_tokenize(texto.lower(), language=idioma)

    # 3. Filtrar palabras que NO estén en stopwords
    palabras_filtradas = [p for p in palabras if p not in stop_words and p.isalpha()]

    return palabras_filtradas

# Ejemplos de uso
textos = [
    "El procesamiento del lenguaje natural es fascinante y útil",
    "La inteligencia artificial está transformando el mundo",
    "Los algoritmos de aprendizaje automático son muy poderosos"
]

print("Remoción de Stopwords:")
print("=" * 60)
for texto in textos:
    resultado = remover_stopwords(texto)
    print(f"Original:      {texto}")
    print(f"Sin stopwords: {' '.join(resultado)}")
    print(f"Palabras removidas: {len(texto.split()) - len(resultado)}")
    print()
```

**Objetivo:** Comparar dos técnicas de reducción de palabras.

**Instrucciones:**
1. Implementa stemming usando SnowballStemmer
2. Implementa lemmatización
3. Compara resultados para las mismas palabras
4. Analiza ventajas y desventajas de cada enfoque

```python
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import WordNetLemmatizer
import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')

def comparar_stemming_lemmatizacion(palabras):
    """
    Compara stemming y lemmatización para un conjunto de palabras.
    """

    stemmer = SnowballStemmer('spanish')
    lemmatizer = WordNetLemmatizer()

    resultados = []

    for palabra in palabras:
        stem = stemmer.stem(palabra)
        lema = lemmatizer.lemmatize(palabra)

        resultados.append({
            'original': palabra,
            'stemming': stem,
            'lemmatizacion': lema
        })

    return resultados

# Palabras en español para stemming
palabras_es = [
    'correr', 'corriendo', 'corrió', 'corrida',
    'programar', 'programación', 'programador',
    'análisis', 'analizar', 'analizando'
]

print("Comparación: Stemming (Español)")
print("=" * 60)
stemmer = SnowballStemmer('spanish')

for palabra in palabras_es:
    print(f"{palabra} -> {stemmer.stem(palabra)}")

# Palabras en inglés para lemmatización
palabras_en = [
    'running', 'runs', 'ran',
    'programming', 'programs', 'programmed',
    'analyzing', 'analyzed', 'analyzes'
]

print("\nComparación: Lemmatización (Inglés)")
print("=" * 60)
lemmatizer = WordNetLemmatizer()

for palabra in palabras_en:
    print(f"{palabra} -> {lemmatizer.lemmatize(palabra, pos='v')}")
```

## **Ejercicio 6: Análisis de Frecuencia de Palabras**

**Objetivo:** Identificar las palabras más frecuentes en un texto.

**Instrucciones:**
1. Implementa función que cuente frecuencia de palabras
2. Remueve stopwords antes de contar
3. Encuentra las N palabras más comunes
4. Visualiza los resultados

```python
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
import re

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

def obtener_palabras_clave(texto, n=10, idioma='spanish'):
    """
    Extrae las N palabras más frecuentes de un texto.
    """

    # 1. Obtener stopwords
    stop_words = set(stopwords.words(idioma))

    # 2. Limpiar texto (minúsculas y sin caracteres especiales)
    texto = texto.lower()
    texto = re.sub(r'[^a-záéíóúüñ\s]', '', texto)

    # 3. Tokenizar
    palabras = word_tokenize(texto, language=idioma)

    # 4. Filtrar stopwords
    palabras_filtradas = [p for p in palabras if p not in stop_words and p.isalpha()]

    # 5. Contar frecuencias
    conteo = Counter(palabras_filtradas)

    # 6. Retornar top N
    return conteo.most_common(n)

# Texto de ejemplo
texto_analisis = """
La Inteligencia Artificial es revolucionaria. La IA transforma industrias.
El aprendizaje automático es clave. Los modelos de IA mejoran constantemente.
Los algoritmos de machine learning procesan datos masivos.
La IA encontrada en aplicaciones modernas es muy sofisticada.
Programación y IA son futuro del software.
"""

print("Análisis de Palabras Clave")
print("=" * 60)

resultado = obtener_palabras_clave(texto_analisis, n=10)

total_palabras = len(texto_analisis.split())
print(f"Total de palabras en el texto: {total_palabras}\n")

print(f"{'Palabra':<15} {'Frecuencia':<12} {'Porcentaje':<10}")
print("-" * 40)

# Imprimir resultados en tabla
for palabra, frecuencia in resultado:
    porcentaje = (frecuencia / total_palabras) * 100
    print(f"{palabra:<15} {frecuencia:<12} {porcentaje:.2f}%")
```

## **Ejercicio 8: Similaridad entre Textos**

**Objetivo:** Calcular qué tan similares son dos textos usando similitud coseno.

**Instrucciones:**
1. Implementa función para calcular similitud con TF-IDF y cosine_similarity
2. Compara diferentes pares de textos
3. Interpreta los resultados (0 = nada similar, 1 = idéntico)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def calcular_similitud_textos(textos):
    """
    Calcula matriz de similitud entre textos usando TF-IDF.

    Args:
        textos: Lista de textos a comparar

    Returns:
        Matriz de similitud (numpy array)
    """
    # 1. Crear vectorizador TF-IDF
    vectorizador = TfidfVectorizer()

    # 2. Ajustar y transformar textos
    matriz_tfidf = vectorizador.fit_transform(textos)

    # 3. Calcular similitud coseno
    matriz_similitud = cosine_similarity(matriz_tfidf)

    # 4. Retornar matriz
    return matriz_similitud

# Textos de prueba
textos_similitud = [
    "El procesamiento del lenguaje natural es importante para la IA",
    "NLP y procesamiento de texto son relevantes en inteligencia artificial",
    "Los gatos son animales domésticos muy populares",
    "La comprensión del lenguaje natural mediante máquinas es fundamental",
    "Los perros y gatos son mascotas comunes en los hogares"
]

print("Análisis de Similitud entre Textos")
print("=" * 80)
print("\nTextos a comparar:")
for i, texto in enumerate(textos_similitud, 1):
    print(f"{i}. {texto}")

print("\n" + "=" * 80)
print("Matriz de Similitud (Similitud Coseno):")
print("=" * 80)

# Calcular la matriz de similitud
matriz = calcular_similitud_textos(textos_similitud)

# Imprimir encabezado de columnas
print("     ", end="")
for i in range(len(textos_similitud)):
    print(f"T{i+1:>7}", end="")
print()

# Imprimir filas con formato legible
for i, fila in enumerate(matriz):
    print(f"T{i+1:>3} ", end="")
    for valor in fila:
        print(f"{valor:>7.2f}", end="")
    print()
```

## **Ejercicio 9: Pipeline Completo de Preprocesamiento**

**Objetivo:** Crear una clase que integre todas las técnicas anteriores.

**Instrucciones:**
1. Implementa clase `PreprocessadorTexto` con métodos para cada paso
2. Integra: limpieza, tokenización, normalización, stopwords, stemming
3. Crea método `procesar()` que execute el pipeline completo
4. Prueba con varios textos complejos

```python
import re
import unicodedata
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer

nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)

class PreprocessadorTexto:
    """
    Pipeline completo de preprocesamiento de texto en español.

    Pasos:
    1. Limpieza (URLs, emails, caracteres especiales)
    2. Normalización (acentos)
    3. Tokenización
    4. Remoción de stopwords
    5. Stemming
    """

    def __init__(self, idioma='spanish'):
        """
        Inicializa el preprocesador.

        Args:
            idioma: Idioma para stopwords y stemming
        """
        self.idioma = idioma
        self.stemmer = SnowballStemmer(idioma)
        self.stopwords_set = set(stopwords.words(idioma))

    def _limpiar(self, texto):
        """Limpia el texto."""
        texto = texto.lower()
        texto = re.sub(r'https?://\S+|www\.\S+', '', texto)
        texto = re.sub(r'\S+@\S+', '', texto)
        texto = re.sub(r'[^a-záéíóúüñ\s]', ' ', texto)
        texto = re.sub(r'\s+', ' ', texto).strip()
        return texto

    def _normalizar(self, texto):
        """Normaliza acentos."""
        texto = unicodedata.normalize('NFD', texto)
        texto = ''.join(
            c for c in texto
            if unicodedata.category(c) != 'Mn'
        )
        return texto

    def _tokenizar(self, texto):
        """Tokeniza el texto en palabras."""
        return word_tokenize(texto, language=self.idioma)

    def _remover_stopwords(self, tokens):
        """Remueve stopwords."""
        return [token for token in tokens if token not in self.stopwords_set and token.isalpha()]

    def _aplicar_stemming(self, tokens):
        """Aplica stemming a cada token."""
        return [self.stemmer.stem(token) for token in tokens]

    def procesar(self, texto, verbose=False):
        """
        Ejecuta el pipeline completo de preprocesamiento.

        Args:
            texto: Texto a procesar
            verbose: Si True, imprime pasos intermedios

        Returns:
            Lista de tokens procesados
        """
        if verbose:
            print(f"Original:\n{texto[:100]}...\n")

        texto_limpio = self._limpiar(texto)
        if verbose:
            print("Después de limpiar:")
            print(texto_limpio[:100], "\n")

        texto_normalizado = self._normalizar(texto_limpio)
        if verbose:
            print("Después de normalizar:")
            print(texto_normalizado[:100], "\n")

        tokens = self._tokenizar(texto_normalizado)
        if verbose:
            print("Después de tokenizar:")
            print(tokens[:20], "\n")

        tokens = self._remover_stopwords(tokens)
        if verbose:
            print("Después de remover stopwords:")
            print(tokens[:20], "\n")

        tokens = self._aplicar_stemming(tokens)

        if verbose:
            print(f"Tokens finales ({len(tokens)} tokens):")
            print(tokens[:20])

        return tokens


# Casos de uso
textos_complejos = [
    """
    ¡Hola a todos! Este es un ejemplo AVANZADO de procesamiento del lenguaje natural.
    Visita https://ejemplo.com o escribe a contacto@empresa.com para más información.
    El NLP es fascinante... ¿No lo crees? #AI #IA @usuario
    """,
    """
    La inteligencia artificial (IA) está revolucionando el mundo!!!
    Los algoritmos de Machine Learning procesan MILLONES de datos.
    La programación en Python3.9+ es excelente (10/10). ¿Verdad???
    """
]

print("Pipeline Completo de Preprocesamiento")
print("=" * 80)

for i, texto in enumerate(textos_complejos, 1):
    print(f"\n\n{'='*80}")
    print(f"EJEMPLO {i}")
    print(f"{'='*80}")

    preprocesador = PreprocessadorTexto()
    tokens_finales = preprocesador.procesar(texto, verbose=True)

    print("\nResultado final:")
    print(tokens_finales)
```

## **Ejercicio 10: Proyecto Integrador - Análisis de Sentimientos Básico**

**Objetivo:** Integrar todas las técnicas para un proyecto simple de análisis de sentimientos.

**Instrucciones:**
1. Preprocesa textos de opiniones
2. Calcula similitud con palabras clave positivas y negativas
3. Clasifica sentimientos (positivo/negativo/neutral)
4. Extrae palabras clave por sentimiento

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from collections import Counter
import re

class AnalizadorSentimientos:
    """
    Analizador simple de sentimientos basado en similitud de textos.
    """

    def __init__(self):
        """
        Inicializa con palabras clave de sentimientos.
        """
        self.palabras_positivas = [
            "excelente", "encanta", "recomendado", "increíble",
            "fantástico", "bueno", "maravilloso", "genial", "superó"
        ]
        self.palabras_negativas = [
            "malo", "terrible", "decepcionante", "frustrante",
            "horrible", "defectuoso", "decepción", "peor", "mal"
        ]
        self.palabras_neutras = [
            "normal", "regular", "tiempo", "producto",
            "llegó", "sorpresas", "común", "estándar"
        ]

    def preprocesar(self, texto):
        """
        Preprocesa el texto: minúsculas, sin caracteres especiales.
        """
        texto = texto.lower()
        texto = re.sub(r'[^a-záéíóúüñ\s]', '', texto)
        texto = re.sub(r'\s+', ' ', texto).strip()
        return texto

    def extraer_palabras_clave(self, texto, n=5):
        """
        Extrae palabras más frecuentes del texto.
        """
        palabras = self.preprocesar(texto).split()
        conteo = Counter(palabras)
        return conteo.most_common(n)

    def analizar(self, texto):
        """
        Analiza sentimiento de un texto.
        """
        texto_limpio = self.preprocesar(texto)

        corpus = [
            texto_limpio,
            " ".join(self.palabras_positivas),
            " ".join(self.palabras_negativas),
            " ".join(self.palabras_neutras)
        ]

        vectorizador = TfidfVectorizer()
        matriz_tfidf = vectorizador.fit_transform(corpus)

        similitud_positiva = cosine_similarity(matriz_tfidf[0:1], matriz_tfidf[1:2])[0][0]
        similitud_negativa = cosine_similarity(matriz_tfidf[0:1], matriz_tfidf[2:3])[0][0]
        similitud_neutral = cosine_similarity(matriz_tfidf[0:1], matriz_tfidf[3:4])[0][0]

        similitudes = {
            "positivo": similitud_positiva,
            "negativo": similitud_negativa,
            "neutral": similitud_neutral
        }

        sentimiento = max(similitudes, key=similitudes.get)
        confianza = similitudes[sentimiento]

        resultado = {
            'texto': texto,
            'sentimiento': sentimiento,
            'confianza': round(confianza, 4),
            'similitud_positiva': round(similitud_positiva, 4),
            'similitud_negativa': round(similitud_negativa, 4),
            'similitud_neutral': round(similitud_neutral, 4),
            'palabras_clave': self.extraer_palabras_clave(texto)
        }
        return resultado

# Textos de prueba
opiniones = [
    "Este producto es excelente y me encanta. Muy recomendado!",
    "Terriblemente malo. No funcionó. Decepcionante y frustrante.",
    "El producto llegó a tiempo. Normal, sin sorpresas.",
    "¡Increíble! Superó todas mis expectativas. Fantástico!",
    "Horrible, defectuoso y de mala calidad. Totalmente decepción."
]

print("Análisis de Sentimientos")
print("=" * 80)

analizador = AnalizadorSentimientos()

print(f"{'Opinión':<55} {'Sentimiento':<12} {'Confianza':<10}")
print("-" * 80)

for opinion in opiniones:
    resultado = analizador.analizar(opinion)
    print(f"{opinion[:50]:<55} {resultado['sentimiento']:<12} {resultado['confianza']:<10}")

print("\nDetalle de análisis:")
print("=" * 80)

for opinion in opiniones:
    resultado = analizador.analizar(opinion)
    print(f"\nTexto: {resultado['texto']}")
    print(f"Sentimiento: {resultado['sentimiento']}")
    print(f"Confianza: {resultado['confianza']}")
    print(f"Similitud positiva: {resultado['similitud_positiva']}")
    print(f"Similitud negativa: {resultado['similitud_negativa']}")
    print(f"Similitud neutral: {resultado['similitud_neutral']}")
    print(f"Palabras clave: {resultado['palabras_clave']}")
```
